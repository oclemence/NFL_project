```python
elo['Winner'] = (
    np.where(
        elo['score1'] > elo['score2'], 
        1, 
        np.where(elo['score1'] < elo['score2'], 0, 0)
    )
)
elo.tail(20)
```

```python
reg_season_mask = elo['playoff'].isnull()
elo_2021 = elo[reg_season_mask]
elo_2021 = elo_2021.drop(['quality', 'importance', 'total_rating', 'neutral'], axis = 1)
elo_2021
```

```python
elo_2021['Elo-difference'] = elo_2021['elo1_pre'] - elo_2021['elo2_pre'] 
elo_2021['QB_Elo-difference'] = elo_2021['qb1_value_pre'] - elo_2021['qb2_value_pre'] 
elo_2021
```

```python
elo_2021['Game_id'] = elo_2021['team1'] + elo_2021['team2']
def sorter(ht):
    words = []
    for s in ht:
        words.append(s)
    words.sort()
    words = ''.join(words)
    return words
elo_2021['Game_id'] = elo_2021['Game_id'].apply(lambda x:sorter(x))
elo_2021['Game_id'] = elo_2021.groupby('Game_id').Game_id.apply(lambda n: n + (np.arange(len(n))+1).astype(str))
elo_2021
```

```python
stats = pd.read_csv('../DF/nflstats_2021.csv', header = 0)
stats
```

```python
elo_2021 = pd.merge(elo_2021, stats, on="Game_id", how="left")
elo_2021
```

```python
elo_2021 = elo_2021.drop(['Unnamed: 0', 'team1_y', 'team2_y'], axis = 1)
elo_2021
```

```python
elo_2021.to_csv('../DF/elo_2021.csv') 
```
