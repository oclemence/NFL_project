```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

main_df = pd.read_csv("../Data/nfl_pass_rush_receive_raw_data.csv", header = 0)
main_df.drop(['player_id', 'Vegas_Line', 'Vegas_Favorite', 'Over_Under'], inplace=True, axis=1)
main_df
```

```python
main_df['game_date'] = pd.to_datetime(main_df.game_date)
start_21 = '9/9/2021'
end_21 = '1/9/2022'
mask = (main_df['game_date'] >= start_21) & (main_df['game_date'] <= end_21)
df_2021 = main_df.loc[mask]
```

```python
start_20 = '9/10/2020'
end_20 = '1/3/2021'
mask = (main_df['game_date'] >= start_20) & (main_df['game_date'] <= end_20)
df_2020 = main_df.loc[mask]
df_2020
```

```python
temps_2020 = []
for s in df_2020.game_id.unique():
    mask = df_2020['game_id'] == s
    masked_df = df_2020[mask]
    temps_2020.append(int(masked_df.iloc[[0]].Temperature))
print(temps_2020)
```

```python
winds_2020 = []
for s in df_2020.game_id.unique():
    mask = df_2020['game_id'] == s
    masked_df = df_2020[mask]
    winds_2020.append(int(masked_df.iloc[[0]].Wind_Speed))
print(len(winds_2020))
```

```python
home_tem2020 = []
home_team2020 = []
for s in df_2020.game_id.unique():
    mask = df_2020['game_id'] == s
    masked_df = df_2020[mask]
    home_tem2020.append(str(masked_df.iloc[[0]].home_team))
for s in home_tem2020:
    t = s.split('\n')[0]
    res = [''.join(g) for _, g in groupby(t, str.isalpha)]
    home_team2020.append(res[1])
print(home_team2020)
```

```python
from itertools import groupby
surf_2020 = []
surface_2020 = []
for s in df_2020.game_id.unique():
    mask = df_2020['game_id'] == s
    masked_df = df_2020[mask]
    surf_2020.append(str(masked_df.iloc[[0]].Surface))
for s in surf_2020:
    t = s.split('\n')[0]
    res = [''.join(g) for _, g in groupby(t, str.isalpha)]
    surface_2020.append(res[1])
print(len(surface_2020))
```

```python
tems = []
for s in df_2020.game_id.unique():
    mask = df_2020['game_id'] == s
    masked_df = df_2020[mask]
    tems.append(str(masked_df.team.unique()))
team1 = []
team2 = []
for t in tems:
    p = t.split("'")
    x = str(p[1])
    y = str(p[3])
    team1.append(x)
    team2.append(y)
print(team2)
```

```python
game_date = []
for s in df_2020.game_id.unique():
    mask = df_2020['game_id'] == s
    masked_df = df_2020[mask]
    print(masked_df.iloc[[0]].game_date)
print(game_date)
```

```python
newDF = pd.DataFrame()
```

```python
newDF['team1'] = team1
newDF['team2'] = team2
newDF['Temp'] = temps_2020
newDF['Wind_speed'] = winds_2020
newDF['surface'] = surface_2020
newDF['home_team'] = home_team2020
newDF.head(20)
```

```python
newDF['team1_home_A'] = (
    np.where(
    newDF['team1'] == newDF['home_team'], 1,
       0)
)
newDF.drop('home_team', axis = 1, inplace = True)
newDF
```

```python
newDF['Game_id'] = newDF['team1'] + newDF['team2'] 
def sorter(ht):
    words = []
    for s in ht:
        words.append(s)
    words.sort()
    words = ''.join(words)
    return words
newDF['Game_id'] = newDF['Game_id'].apply(lambda x:sorter(x))
newDF
```

```python
newDF['Game_id'] = newDF.groupby('Game_id').Game_id.apply(lambda n: n + (np.arange(len(n))+1).astype(str))
newDF
```
