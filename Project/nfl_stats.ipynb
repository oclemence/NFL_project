```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

main_df = pd.read_csv("/Users/oliverclemence/Downloads/nfl_pass_rush_receive_raw_data.csv", header = 0)
main_df.drop(['player_id', 'Vegas_Line', 'Vegas_Favorite', 'Over_Under'], inplace=True, axis=1)
main_df
```

```python
stats = {}
for s in main_df.game_id.unique():
    mask = main_df['game_id'] == s
    game_s = []
    game_s.append('Temp: ' + str(main_df[mask].Temperature.unique()))
    game_s.append(str(main_df[mask].home_team.unique()))
    game_s.append(str(main_df[mask].vis_team.unique()))
    game_s.append(str(main_df[mask].game_date.unique()))
    game_s.append('Humidity: ' + str(main_df[mask].Humidity.unique()))
    game_s.append('Wind_speed: ' + str(main_df[mask].Wind_Speed.unique()))
    stats[str(main_df[mask].home_team.unique()) +str(main_df[mask].vis_team.unique())] = game_s
print(stats)
```

```python
def parse_ht(season):
    ht_ = season.split("-")
    return(ht_[0])
def parse_m(season):
    m = season.split("-")
    return(int(m[1]))
def parse_y(season):
    y = season.split("-")
    return(int(y[2]))

main_df["year"] = main_df['game_date'].apply(lambda x:parse_ht(x))
main_df["month"] = main_df['game_date'].apply(lambda x:parse_m(x))
main_df["day"] = main_df['game_date'].apply(lambda x:parse_y(x))

main_df
mask = ((main_df['year'].isin(['2020']) & main_df['month'].isin([9,10,11,12])))
mask1 = main_df[mask]
mask2 = (main_df['year'].isin(['2021']) & main_df['month'].isin([1]))
mask2 = main_df[mask2]
df_2020 = mask1.append(mask2)
mask3 = df_2020['month'] == 1 & df_2020['day'].isin(range(5,31))
df_2020 = df_2020[~mask3]
df_2020
```

```python
mask = ((main_df['year'].isin(['2021']) & main_df['month'].isin([9,10,11,12])))
mask1 = main_df[mask]
mask2 = (main_df['year'].isin(['2022']) & main_df['month'].isin([1]))
mask2 = main_df[mask2]
df_2021 = mask1.append(mask2)
mask3 = df_2021['month'] == 1 & df_2021['day'].isin(range(10,31))
df_2021 = df_2021[~mask3]
df_2021
```

```python
temps_2020 = []
for s in df_2020.game_id.unique():
    mask = df_2020['game_id'] == s
    masked_df = df_2020[mask]
    temps_2020.append(int(masked_df.iloc[[0]].Temperature))
print(temps_2020)
```

```python
winds_2020 = []
for s in df_2020.game_id.unique():
    mask = df_2020['game_id'] == s
    masked_df = df_2020[mask]
    winds_2020.append(int(masked_df.iloc[[0]].Wind_Speed))
print(len(winds_2020))
```

```python
surface_2020 = []
for s in df_2020.game_id.unique():
    mask = df_2020['game_id'] == s
    masked_df = df_2020[mask]
    surface_2020.append(masked_df.iloc[[0]].Surface)
print(len(surface_2020))
```

```python
month_2020 = []
for s in df_2020.game_id.unique():
    mask = df_2020['game_id'] == s
    masked_df = df_2020[mask]
    month_2020.append(int(masked_df.iloc[[0]].month))
print(month_2020)
```

```python
day_2020 = []
for s in df_2020.game_id.unique():
    mask = df_2020['game_id'] == s
    masked_df = df_2020[mask]
    day_2020.append(int(masked_df.iloc[[0]].day))
print(day_2020)
```

```python
tems = []
for s in df_2020.game_id.unique():
    mask = df_2020['game_id'] == s
    masked_df = df_2020[mask]
    tems.append(str(masked_df.team.unique()))
team1 = []
team2 = []
for t in tems:
    p = t.split("'")
    x = str(p[1])
    y = str(p[3])
    team1.append(x)
    team2.append(y)
print(team2)
```

```python
game_date = []
for s in df_2020.game_id.unique():
    mask = df_2020['game_id'] == s
    masked_df = df_2020[mask]
    print(masked_df.iloc[[0]].game_date)
print(game_date)
```
